```python
from matplotlib import pyplot as plt
import math as mt

x = range(1, 11)

y1 = [0*n + 1 for n in x]
y2 = [mt.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * mt.log(n) for n in x]
y5 = [n ** 2 for n in x]
y6 = [2 ** n for n in x]
y7 = [mt.factorial(n) for n in x]
y8 = [n ** n for n in x]

plt.figure(figsize=(16, 9))

plt.plot(x, y1, label='O(1)')
plt.plot(x, y2, label='O(log n)')
plt.plot(x, y3, label='O(n)')
plt.plot(x, y4, label='O(n log n)')
plt.plot(x, y5, label='O(n^2)')
plt.plot(x, y6, label='O(2^n)')
plt.plot(x, y7, label='O(n!)')
plt.plot(x, y8, label='O(n**n)')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.style.use('dark_background')
plt.show()
```
