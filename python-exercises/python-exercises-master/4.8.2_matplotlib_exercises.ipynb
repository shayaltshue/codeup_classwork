```python
from matplotlib import pyplot as plt
import math as mt
```

1. Use matplotlib to plot the following equation: y = x^2 - x + 2
    You'll need to write the code that generates the x and y points.
    Add an anotation for the point 0, 0, the origin.

```python
# y = x^2 - x + 2

x = range(1, 25)
y = [n**2 - n + 2 for n in x]

plt.title('$y = x^2 - x + 2$')
plt.annotate("Origin", xy=(0,0), xytext=(0, 125), arrowprops={'facecolor': 'blue'})
plt.plot(y)
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

```python
# 2a. y = √x

x = range(1, 11)
y = [mt.sqrt(n) for n in x]

plt.title('$y = √x$')
plt.plot(y)
plt.show()
```

```python
# 2b. y = x^3

x = range(1, 10)
y = [n ** 3 for n in x]

plt.title('$y = x^3$')
plt.plot(x, y)
plt.show()
```

```python
# 2c. y = tan(x)

x = range(1, 11)
y = [mt.tan(n) for n in x]

plt.title('$y = tan(x)$')
plt.plot(x, y)
plt.show()
```

```python
# 2d. y = 2^x

x = range(1, 11)
y = [2 ** n for n in x]

plt.title('$y = 2^x$')
plt.plot(y)
plt.show()
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.


# subplot 1 (aka 2a)
x = range(1, 11)
y1 = [mt.sqrt(n) for n in x]

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('square root of n')

# subplot 2 (aka 2b)
y2 = [n ** 3 for n in x]

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('n to the power of 3')

# subplot 3 (aka 2c)
y3 = [mt.tan(n) for n in x]

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('tan of n')

# subplot 4 (aka 2D)
y4 = [2 ** n for n in x]

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('2 to the power of n')
```

```python
# 4Comine the figures you created in the last step into one figure 
# where each of the 4 equations has a different color for the points. 
# Be sure to include a legend.

x = range(1, 11)
y1 = [mt.sqrt(n) for n in x]
y2 = [n ** 3 for n in x]
y3 = [mt.tan(n) for n in x]
y4 = [2 ** n for n in x]

plt.plot(x, y1, color='purple', label='square root of n')
plt.plot(x, y2, color='k', label='n to the power of 3')
plt.plot(x, y3, color='blue', linestyle= 'dashed', label='tan of n')
plt.plot(x, y4, color='pink', label='2 to the power of n')
plt.legend()
```
